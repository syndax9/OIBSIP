```python
#First we will import all the required libraries for the analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Collecting our dataset
data_before_covid = pd.read_csv("Unemployment in India.csv")
```

```python
data_before_covid.shape
```

```python
#These are all the unique Regeion present in our dataset
data_before_covid.Region.unique()
```

```python
#This cell output contains all the necessary information about our datset like all the columns with indexes and NULL columns
data_before_covid.describe
```

```python
#Here it shows all the null columns present in our dataset
data_before_covid.isnull().sum()
```

```python
#The number of NULL cells are quite low comparitively 
data_before_covid = data_before_covid.dropna()
```

```python
data_before_covid.shape
```

```python
x=data_before_covid['Region']
y=data_before_covid[' Estimated Unemployment Rate (%)']
```

### This is a histogram which shows our Estimated unemployment rate according to different states 

```python
plt_1_unemployed = plt.figure(figsize=(40, 15))
plt.title('Umemployment Rate', fontweight='bold' ,fontsize=18)

plt.xlabel("Region",fontweight='bold',fontsize=20)
plt.ylabel("Estimated Unemployment rate",fontweight='bold',fontsize=18)
plt.xticks(fontweight='bold',rotation='vertical',fontsize=18)
plt.bar(x,y, color='b',align='center')
```

```python
a=data_before_covid[' Estimated Employed']
b=data_before_covid['Region']
```

### This line plot shows our Employed numbers with respect to states 

```python
plt_1_employed = plt.figure(figsize=(5, 15))
plt.plot(a, b)
```

### This shows the correlation between different columns in our dataset 

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(12, 10))
sns.heatmap(data_before_covid.corr())
plt.show()
```

```python
p=data_after_covid[' Estimated Unemployment Rate (%)']
q=data_after_covid['Region']
```

### This bar graph shows unemployment rate according to states 

```python
plt_2_unemployed = plt.figure(figsize=(40, 15))
plt.title('Umemployment Rate', fontweight='bold' ,fontsize=20)

plt.xlabel("States",fontweight='bold',fontsize=20)
plt.ylabel("Estimated Unemployment rate",fontweight='bold',fontsize=20)
plt.xticks(fontweight='bold',rotation='vertical',fontsize=20)
plt.bar(q,p, color='r',align='center')
```

```python
fig_after_covid = plt.figure(figsize = (30, 15))
plt.scatter(data_after_covid[' Date'], data_after_covid[' Estimated Employed'])
plt.title('Unemployment according to Region')
plt.xlabel('Region')
plt.ylabel('Unemployment Rate')
plt.show()
```

```python
m = data_after_covid[' Estimated Employed']
n = data_after_covid['Region']
```

### this line plot shows our Employed numbers with respect to states
